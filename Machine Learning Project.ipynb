```python
def extract_fname(name):
    #splitted the data at dot
    z=name.split(".")
    #second element of z will be stripped and splitted 
    a=z[1].strip()
    b=a.split(" ")
    fname=b[0]
    return(fname)
```

```python
extract_fname(text)
```

```python
ship.Name.apply(extract_fname)
```
